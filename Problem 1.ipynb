```python
import pandas as pd
df = pd.read_csv("hello.csv")
df
```

```python
df.describe()
```

```python
Q1 = df.order_amount.quantile(0.25)
Q3 = df.order_amount.quantile(0.75)
Q1, Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit, upper_limit
```

```python
df[(df.order_amount<lower_limit)|(df.order_amount>upper_limit)]
```

```python
df_no_outlier = df[(df.order_amount>lower_limit)&(df.order_amount<upper_limit)]
df_no_outlier
```

```python
df_no_outlier.order_amount.mean()
```
